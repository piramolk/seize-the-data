# Pandas
## 1. Basic size and shape
* df.head()/tail : to see what it looks like
* df.columns : get all column names in case there are too many to be seen
* df.shape : get the dimensions
    * helps determine dependent and independent variales
* df.info()
    * good to know dtypes in case need to convert
         * if you do need, do e.g. `df['a'].astype('int')` if u just wanna print or  `df['a'] = df['a'].astype('int')' to change
         * use `df['a'].dtype` to doublecheck
* df.describe()
    * get summary stats
* if you wanna be more manual
    * `df.sum()` : sum of values
    * `df.cumsum()` : cumulative sum of values
    * `df.min()` , `df.max()` : min/max vals
    *  `df.idxmin()/df.idxmax()`: Minimum/Maximum index value
    * `df.mean()` Mean of values
    * `df.median()` Median of values
    
* `df['value'].value_counts()` or `missing_value_effect['value'].value_counts(dropna=False)`

# Analyzing relationships between variables

## 1. Correlation Matrix
* To review, correlation is a measurement that describes the relationship between two variable
* Thus, a correlation matrix is a table that shows the correlation coefficients between many variables. I used sns.heatmap() to plot a correlation matrix of all of the variables in the used car dataset.

```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
# calculate correlation matrix
corr = df.corr()# plot the heatmap
plt.figure(figsize=(10,10))
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

## 2. Scatterplot

> A scatterplot is a type of graph which â€˜plotsâ€™ the values of two variables along two axes, like age and height. 
* Scatterplots are useful for many reasons: like correlation matrices, it allows you to quickly understand a relationship between two variables, itâ€™s useful for identifying outliers, and itâ€™s instrumental when polynomial multiple regression models 
* I used .plot() and set the â€˜kindâ€™ of graph as scatter. I also set the x-axis to â€˜odometerâ€™ and y-axis as â€˜priceâ€™, since we want to see how different levels of mileage affects price.

```python
df.plot(kind='scatter', x='Distance', y='Price')
```

`sns.pairplot(df)`


## 3. Histogram

* Correlation matrices and scatterplots are useful for exploring the relationship between two variables. But what if you only wanted to explore a single variable by itself? 
* This is when histograms come into play. Histograms look like bar graphs but they show the distribution of a variableâ€™s set of values.


```python
df['Price'].plot(kind='hist', bins=50, figsize=(12,6), facecolor='grey',edgecolor='black')
```

* positive skew (right tail)
* Skewness is important, especially in areas like finance, because a lot of models assume that all variables are normally distributed, which typically isnâ€™t the case.
